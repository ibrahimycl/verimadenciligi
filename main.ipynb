```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("movie/titles.csv")

df['imdb_score'] = df['imdb_score'].fillna(0) 

yearly_imdb_avg = df.groupby('release_year')['imdb_score'].mean().reset_index()

# Seaborn ile çizim
plt.figure(figsize=(14, 6))
sns.barplot(x='release_year', y='imdb_score', data=yearly_imdb_avg, palette='viridis')
plt.xlabel('Yıl')
plt.ylabel('IMDb Ortalaması')
plt.title('Her Yılın IMDb Ortalaması')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```
